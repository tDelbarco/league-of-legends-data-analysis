## archivos

```python
from pathlib import Path

ruta = Path(r"dragontail-16.1.1\16.1.1\data\es_AR")

archivos = []

for e in ruta.iterdir():
    print(e)
    archivos.append(e)
```

## tft

```python
tft = []

for e in archivos:
    if e.is_file() and e.name.startswith("tft"):
        tft.append(e)

for e in tft:
    archivos.remove(e)
```

##  carpetas

```python
carpetas = [e for e in archivos if e.is_dir()]

for d in carpetas:
    archivos.remove(d)

carpetas
```
