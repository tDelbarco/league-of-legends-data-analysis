```python
def inspeccionar_json(diccionario):
    tipos_simples = (str, int, float, bool)

    for clave, valor in diccionario.items():
        if isinstance(valor, tipos_simples):
            # Si es simple, mostramos clave : valor
            print(f"{clave} : {valor}")
        else:
            # Si es complejo (dict o list), mostramos clave : tipo
            print(f"{clave} : <Tipo: {type(valor).__name__.upper()}>")
```

```python
for e in champion_paths:
    with e.open(encoding="utf-8") as f:
        data = json.load(f)
        key = list(data.get("data").keys())[0]
        data = data.get("data").get(key)
        #print(data)
        for spell in data.get("spells"): #habilidades
            items = list(spell.items())
            items.insert(0, ('championId', data.get("id")))
            #print(dict(items))
            df_temp = pd.json_normalize(dict(items))
            df_temp.columns = [c.replace('.', '_') for c in df_temp.columns]
            columnas_interes = ['id','championId', 'name', 'description', 'maxrank', 'cooldown', 'effect', 'range', 'image_full', 'resource']
            df_final = df_temp[columnas_interes].rename(columns={'id': 'spellId'})
            try:
                df_final.to_sql(
                    'champions_skills', 
                    engine, 
                    if_exists='append', 
                    index=False,
                    dtype={
                        'cooldown': JSONB,
                        'effect': JSONB,
                        'range': JSONB
                    }
                )
            except Exception as e:
                print(f"Error al insertar habilidad {df_final['spellId'].iloc[0]} del campeón {df_final['championId'].iloc[0]}: {e}")
        df = pd.DataFrame([data])
        df.drop(columns=['spells'], inplace=True) #ya que spells se tranformo en otra tabla
        df_flatten = pd.json_normalize(data)

        columnas_con_punto = [col for col in df_flatten.columns if '.' in col]

        # Extraemos la raíz (lo que está antes del primer punto) y sacamos únicos
        raices_anidadas = pd.Series(columnas_con_punto).str.split('.').str[0].unique().tolist()
        df_flatten.columns = [c.replace('.', '_') for c in df_flatten.columns]
        
        df = df.rename(columns={'id': 'championId'})
        df = df.drop(columns=raices_anidadas, errors='ignore')

        grupos = {raiz: df_flatten.filter(regex=rf'^{raiz}\_') for raiz in raices_anidadas}

        for key in grupos.keys():
            grupos[key].insert(0, 'championId', df['championId'])
            grupos[key].columns = [col.split('.')[-1] for col in grupos[key].columns]

        df = pd.merge(df, grupos["image"], on='championId', how='left')
        df = pd.merge(df, grupos["info"], on='championId', how='left')
        df = pd.merge(df, grupos["passive"], on='championId', how='left')

        print(df.columns)
        
        try:
            df.to_sql(
                'champions', 
                engine, 
                if_exists='append', 
                index=False,
                dtype={
                    'skins': JSONB,
                    'allytips': JSONB,
                    'enemytips': JSONB,
                    'tags': JSONB,
                    'recommended': JSONB
                }
            )
        except Exception as e:
            print(f"Error al insertar campeón {df_final['championId'].iloc[0]}: {e}")

        
```
