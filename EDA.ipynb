```python
def inspeccionar_json(diccionario):
    tipos_simples = (str, int, float, bool)

    for clave, valor in diccionario.items():
        if isinstance(valor, tipos_simples):
            # Si es simple, mostramos clave : valor
            print(f"{clave} : {valor}")
        else:
            # Si es complejo (dict o list), mostramos clave : tipo
            print(f"{clave} : <Tipo: {type(valor).__name__.upper()}>")
```

## tabla champion_skills

```python
class ChampionSkills(Base):
    __tablename__ = "champions_skills"
    
    # PK Compuesta para Skills: un hechizo es único en una versión
    spellId = Column(String, primary_key=True) 
    version = Column(String(20), primary_key=True)

    championId = Column(String(25))
    name = Column(String)
    description = Column(String)
    maxrank = Column(Integer)
    cooldown = Column(JSONB) 
    effect = Column(JSONB)   
    range = Column(JSONB)
    image_full = Column(String)
    resource = Column(String)

    # Relación y FK Compuesta: apunta a un registro exacto de Champion (ID + Versión)
    __table_args__ = (
        ForeignKeyConstraint(
            ['championId', 'version'],
            ['champions.championId', 'champions.version'],
        ),
    )
    champion = relationship("Champion", back_populates="skills")
```

```python
class ChampionStats(Base):
    __tablename__ = "champion_stats"

    # PK Compuesta para Stats: asegura que no se repitan las stats de un campeón en la misma versión
    championId = Column(String(25), primary_key=True)
    version = Column(String(20), primary_key=True)

    stats_hp = Column(Float)
    stats_hpperlevel = Column(Float)
    stats_mp = Column(Float)
    stats_mpperlevel = Column(Float)
    stats_movespeed = Column(Float)
    stats_armor = Column(Float)
    stats_armorperlevel = Column(Float)
    stats_spellblock = Column(Float)
    stats_spellblockperlevel = Column(Float)
    stats_attackrange = Column(Float)
    stats_hpregen = Column(Float)
    stats_hpregenperlevel = Column(Float)
    stats_mpregen = Column(Float)
    stats_mpregenperlevel = Column(Float)
    stats_crit = Column(Float)
    stats_critperlevel = Column(Float)
    stats_attackdamage = Column(Float)
    stats_attackdamageperlevel = Column(Float)
    stats_attackspeedperlevel = Column(Float)
    stats_attackspeed = Column(Float)

    # Relación y FK Compuesta
    __table_args__ = (
        ForeignKeyConstraint(
            ['championId', 'version'],
            ['champions.championId', 'champions.version'],
        ),
    )
    champion = relationship("Champion", back_populates="stats")
```

## tabla champions

```python
class Champion(Base):
    __tablename__ = "champions"

    # PK Compuesta principal
    championId = Column((String), primary_key=True)
    version = Column(String(20), primary_key=True)

    key = Column(String)
    name = Column(String)
    title = Column(String)
    skins = Column(JSONB)
    lore = Column(String)
    blurb = Column(String)
    allytips = Column(JSONB)
    enemytips = Column(JSONB)
    tags = Column(JSONB)
    partype = Column(String)
    recommended = Column(JSONB)

    image_full = Column(String)
    image_sprite = Column(String)
    image_group = Column(String)
    image_x = Column(Integer)
    image_y = Column(Integer)
    image_w= Column(Integer)
    image_h = Column(Integer)

    info_attack = Column(Integer)
    info_defense = Column(Integer)
    info_magic = Column(Integer)
    info_difficulty = Column(Integer)
    
    passive_name = Column(String)
    passive_description = Column(String)
    passive_image_full = Column(String)
    passive_image_sprite = Column(String)
    passive_image_group = Column(String)
    passive_image_x = Column(Integer)
    passive_image_y = Column(Integer)
    passive_image_w = Column(Integer)
    passive_image_h = Column(Integer)

    # Relaciones que usan la PK compuesta automáticamente
    skills = relationship("ChampionSkills", back_populates="champion")
    stats = relationship("ChampionStats", back_populates="champion")
```

```python
Base.metadata.create_all(engine)
```

```python
import pandas as pd
import json
from sqlalchemy.dialects.postgresql import JSONB

for e in champion_paths:
    with e.open(encoding="utf-8") as f:
        data_raw = json.load(f)
        version = data_raw.get("version")
        key_name = list(data_raw.get("data").keys())[0]
        data = data_raw.get("data").get(key_name)

        # --- PROCESAMIENTO DE HABILIDADES ---
        skills_to_insert = []
        for spell in data.get("spells"):
            items = list(spell.items())
            items.insert(0, ('championId', data.get("id")))
            df_temp_skill = pd.json_normalize(dict(items))
            df_temp_skill.columns = [c.replace('.', '_') for c in df_temp_skill.columns]
            
            columnas_interes = ['id', 'championId', 'name', 'description', 'maxrank', 'cooldown', 'effect', 'range', 'image_full', 'resource']
            df_skill = df_temp_skill[columnas_interes].rename(columns={'id': 'spellId'})
            df_skill.insert(1, 'version', version)
            skills_to_insert.append(df_skill)

        # --- PROCESAMIENTO DE CAMPEÓN Y STATS ---
        df_champ = pd.DataFrame([data])
        df_champ.drop(columns=['spells'], inplace=True)
        
        # Flatten para obtener stats, image, info, passive
        df_flatten = pd.json_normalize(data)
        columnas_con_punto = [col for col in df_flatten.columns if '.' in col]
        raices_anidadas = pd.Series(columnas_con_punto).str.split('.').str[0].unique().tolist()
        df_flatten.columns = [c.replace('.', '_') for c in df_flatten.columns]
        
        # Limpieza del DF principal de campeón
        df_champ = df_champ.rename(columns={'id': 'championId'})
        df_champ = df_champ.drop(columns=raices_anidadas, errors='ignore')
        df_champ.insert(1, 'version', version)

        # Separación de grupos (stats, info, etc)
        grupos = {raiz: df_flatten.filter(regex=rf'^{raiz}\_') for raiz in raices_anidadas}

        for r_key in grupos.keys():
            grupos[r_key].insert(0, 'championId', df_champ['championId'].iloc[0])

        # Merges para la tabla principal (image, info, passive)
        df_champ = pd.merge(df_champ, grupos["image"], on='championId', how='left')
        df_champ = pd.merge(df_champ, grupos["info"], on='championId', how='left')
        df_champ = pd.merge(df_champ, grupos["passive"], on='championId', how='left')

        # Preparar tabla Stats
        df_stats = grupos["stats"].copy()
        df_stats.insert(1, 'version', version)

        try:
            with engine.begin() as connection:
                # INSERTAR CHAMPIONS
                df_champ.to_sql(
                    'champions', 
                    connection, 
                    if_exists='append', 
                    index=False,
                    dtype={
                        'skins': JSONB,
                        'allytips': JSONB,
                        'enemytips': JSONB,
                        'tags': JSONB,
                        'recommended': JSONB
                    }
                )

                # INSERTAR STATS
                df_stats.to_sql(
                    'champion_stats', 
                    connection, 
                    if_exists='append', 
                    index=False
                )

                # INSERTAR SKILLS
                for df_s in skills_to_insert:
                    df_s.to_sql(
                        'champions_skills', 
                        connection, 
                        if_exists='append', 
                        index=False,
                        dtype={
                            'cooldown': JSONB,
                            'effect': JSONB,
                            'range': JSONB
                        }
                    )
            
            print(f"Campeón {data.get('id')} ({version}) insertado correctamente.")

        except Exception as ex:
            print(f"Error al insertar campeón {data.get('id')}: {ex}")
```
